#concatanet slice Embedding

```python
list_path_tensor = [f'/content/drive/MyDrive/Sentiment Analys Data/review_embbeding_{slice}.pt' for slice in range(100,701,100)]
# load slice embedding review
tensors = [ torch.load(path,map_location=torch.device('cpu')) for path in list_path_tensor]
# cancatenate slice embedding review
review_embedding = torch.cat(tensors, dim =0)
#save embedding review as one tensor
torch.save(review_embedding, '/content/drive/MyDrive/Sentiment Analys Data/review_embbeding')
```

```python
# load embedding review
review_embedding = torch.load('/content/drive/MyDrive/Sentiment Analys Data/review_embbeding'
                                ,map_location=torch.device('cpu'))
```

```python
# normelize for using dot product
review_embedding = util.normalize_embeddings(review_embedding)
```

# finde reviews containing 'guaranty' and' warranty'

```python
guarantee = X.str.lower().str.contains('guarantee')
warranty = X.str.lower().str.contains('warrant')
print(f'The number review contains guarantee :{guarantee.sum()}')
print(f'The number review contains warranty :{warranty.sum()}')
print(f'The number review contains guaranty or warranty :{(warranty|guarantee).sum()}')
```

## Calculate the similarity

```python
query_embedding = review_embedding[X[(warranty|guarantee)].index[0: 100]]
```

```python
similarity = util.semantic_search(query_embedding ,
                                         review_embedding,
                                         score_function = similarity_functions.dot_score,
                                         top_k = 10)
```

```python
import re
ids_list = []
for query in similarity:
  for s in query:
    ids_list.append(s['corpus_id'])
# sum of similar review containing warrany or guarantee
total = X[ids_list].str.contains('warranty|guarantee',case = False, regex = True).sum()

(total-100) / (len(ids_list)-100)
```
